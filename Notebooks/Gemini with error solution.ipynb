```python
import re
import time
from PIL import Image

text = ""
errors = []

# You must define `all_files` earlier with os.walk()
delay_point = list(range(0, len(all_files), 15))  # Every 15 files

for index, file_path in enumerate(all_files):
    if file_path.endswith('.jpg') or file_path.endswith('.jpeg'):
        try:
            image = Image.open(file_path)
            response = model.generate_content(["Extract the text of this image:", image])

            if response.candidates and response.candidates[0].finish_reason != 'OTHER':
                extracted = response.text
            else:
                extracted = ""
                errors.append(file_path)

            # Clean and append text
            extracted = re.sub(r'Here is the text from the image:', '', extracted)
            text += extracted + "\n\n---\n"

        except Exception as e:
            errors.append(file_path)
            print(f"❌ Error processing {file_path}: {e}")
        print(f"{index} of {len(all_files)} processed")

        if n in delay_point:
            print("⏸️ Sleeping for 60 seconds to avoid rate limits...")
            with open("text.txt", "a", encoding="utf-8") as f:
                f.write(text)
            text = ""
            time.sleep(60)

# Final save
print(f" Finished {len(all_files)} files.")
print(f"Remaining errors: {len(errors)}")
with open("text.txt", "a", encoding="utf-8") as f:
    f.write(text)
    text = ""

# Optional: save error list
if errors:
    with open("errors.txt", "x", encoding="utf-8") as f:
        for e in errors:
            f.write(e + "\n")
```

```python
import time
from PIL import Image
import google.generativeai as genai
# One solution to the errors is to try the files again on a loop, assuming the error was caused by something such as copyright misidentification
# If a lower quality model is chosen, this could impact the OCR quality, but not substantially
model_ids = [
    "gemini-2.0-flash",
    "gemini-2.0-flash-lite",
    "gemini-1.5-flash",
    "gemini-1.5-flash-8b"
]

MAX_ROUNDS = 20
text = ""

for round_counter in range(1, MAX_ROUNDS + 1):
  print(f"\n🔁 Round {round_counter}")
  for model_id in model_ids:
        model = genai.GenerativeModel(model_id)
        still_errors = []

        for file_path in errors:
            try:
                image = Image.open(file_path)
                response = model.generate_content(["Extract the text of this image:", image])

                if response.candidates and response.candidates[0].finish_reason != 4:
                    extracted = response.text
                    text += extracted + "\n\n---\n"
                    print(f"✅ {file_path} extracted using {model_id}")
                else:
                    still_errors.append(file_path)

            except Exception:
                still_errors.append(file_path)

            time.sleep(4.2)

        errors = still_errors
        print(f"remaining files {len(errors)}")


  if not errors:
        break
f = open(r"text.txt", "a", encoding="utf-8")
f.write(text)
f.close()
text = ""
```
